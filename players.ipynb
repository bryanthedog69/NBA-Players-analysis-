```python
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("C:/Users/hp/Documents/python_files/Regular_Season.csv")
data.head()
```

PERCENTAGE OF PLAYERS THAT PLAY FOR THE SAME TEAM 

```python
 #Group by TEAM and count unique players
team_player_count = data.groupby('TEAM')['PLAYER'].nunique()
```

```python
#Calculate the percentage of players for each team 
team_player_percentage = (team_player_count / team_player_count.sum()) * 100
```

```python
# Plotting the pie chart using Matplotlib directly
plt.figure(figsize=(10,8))
plt.pie(team_player_percentage, labels=team_player_percentage.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.tab20.colors)
plt.title('Percentage of players playing for the same team')
plt.show()
```
